# Décrire les col de type Categorical : Recherche d'Outlier, value_counts(), ... ?

```python
cat_cols = ['marital','job','education','targeted','default','housing','loan',
            'contact','day','month','poutcome','response']

for col in cat_cols:
    print(f"Column: {col}")
    print("Unique values:", df[col].unique())
    print("Value counts:")
    print(df[col].value_counts())
    print("-"*40)
```
